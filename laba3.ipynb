```python
from dataclasses import dataclass
import yaml
import matplotlib.pyplot as plt

# Импортируем attrs только для объявления атрибута-функции в dataclass
import attr

# Шаг 1: Определяем структуру данных с помощью dataclasses
@dataclass
class DataPoint:
    time: float
    value: float

# Шаг 2: Функция для чтения YAML файла и создания объекта с данными
def read_yaml_file(file_path):
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)
        data_points = [DataPoint(point['time'], point['value']) for point in data]
        return data_points

# Шаг 3: Методы для работы с данными
def plot_time_series(data_points):
    times = [point.time for point in data_points]
    values = [point.value for point in data_points]

    plt.plot(times, values)
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title('Time Series Data')
    plt.show()

# Пример использования
if __name__ == "__main__":
    # Путь к YAML файлу
    yaml_file_path = 'C:\Project\Python\data.yaml'
    
    # Чтение файла и создание объекта с данными
    data_points = read_yaml_file(yaml_file_path)
    
    # Построение графика временного ряда
    plot_time_series(data_points)
```
